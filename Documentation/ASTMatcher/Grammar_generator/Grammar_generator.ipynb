```python
def read_file(file_name):
    return(open(file_name, 'r', encoding='utf-8').readlines())
```

```python
narrowing_api = read_file("narrowing_function.txt")
narrowing_input = read_file('narrowing_input.txt')
narrowing_output = read_file('narrowing_return.txt')
narrowing_desc = read_file('narrowing_description.txt')

traversal_api = read_file('traversal_function.txt')
traversal_input = read_file('traversal_input.txt')
traversal_output = read_file('traversal_return.txt')
traversal_desc = read_file('traversal_description.txt')

node_api = read_file('node_function.txt')
node_input = read_file('node_input.txt')
node_output = read_file('node_return.txt')
node_desc = read_file('node_description.txt')

# print(narrowing_api)
```

```python
node_matcher_list = []
narrowing_matcher_list = []
traversal_matcher_list = []
```

```python
import Grammar as grammar
gram = grammar.Grammar()
for i in range(len(narrowing_api)):
    name = narrowing_api[i].replace('\n', '')
    desc = narrowing_desc[i].replace('\n', '')
    output = narrowing_output[i].replace('\n', '')
    param = narrowing_input[i].replace('\n', '').split(',')
    
    gram.add_api(name, param, output, desc, 'narrowing')
    narrowing_matcher_list.append(name)
    
for i in range(len(node_api)):
    name = node_api[i].replace('\n', '')
    desc = node_desc[i].replace('\n', '')
    output = node_output[i].replace('\n', '')
    param = node_input[i].replace('\n', '').split(',')
    
    gram.add_api(name, param, output, desc, 'node')
    node_matcher_list.append(name)
    
for i in range(len(traversal_api)):
    name = traversal_api[i].replace('\n', '')
    desc = traversal_desc[i].replace('\n', '')
    output = traversal_output[i].replace('\n', '')
    param = traversal_input[i].replace('\n', '').split(',')
    
    gram.add_api(name, param, output, desc, 'traversal')
    traversal_matcher_list.append(name)
    
gram.api_display()
```

```python
# set type class dict  
type_class_dict = {} # {type:[inherent types]}
type_class_dict['decl'] = ['']
type_class_dict['nameddecl'] = []
type_class_dict['templatedecl'] = ['conceptdecl']
type_class_dict['typedecl'] = ['tagDecl', 'typedefnamedecl']
type_class_dict['stmt'] = []
type_class_dict['qualtype'] = ['type']

stmt_text = "asmstmt, attributedstmt, breakstmt, capturedstmt, continuestmt, coreturnstmt, csscatchstmt,\
             cxxforrangestmt, csstrystmt, declstmt, dostmt, expr, forstmt, gotostmt, ifstmt, indirectgotostmt, labelstmt,\
             msdependentexitstsstmt, nullstmt, objcatcatchstmt, objcatfinallystmt, objcatsynchroizedstmt, objcatthrowstmt,\
             objcattrystmt, objcautoreleasepoolstmt, objcforcollectionstmt, ompexecutabledirective, returnstmt, sehexceptstmt,\
             sehfinallystmt, sehleavestmt, sehtrystmt, switchcase, switchstmt, valuestmt, whilestmt,"

decl_text = 'accessspecdecl, blockdecl, captureddecl, classscopefunctionspecializationdecl, emptydecl, exportdecl, externccontextdecl,\
             filescopeasmdecl, frienddecl, friendtemplatedecl, importdecl, linkagespecdecl, nameddecl, objcpropertyimpldecl,\
             omprequiresdecl, ompthreadprivatedecl, pragmacommentdecl, pragmadetectmismatchdecl, staticassertdec, translationunitdecl'
             
nameddecl_text = 'labeldecl, namespacealiasdecl, namespacedecl, objccompatiblealiasdecl, objccontainerdec, objcmethoddecl, objcpropertdecl,\
             templatedecl, typedecl, usingdecl, usingdirectivedecl, usingpackdecl, usingshadowdecl, valuedecl' 

objccontainerdecl_text = 'objccategorydecl, objclmpldecl, objcinterfacedecl, objcprotocoldecl'

templateDecl_text = 'builtintemplatedecl, redeclarabletemplatedecl, templatetemplateparmdecl'

typeDecl_text = 'tagdecl, templatetypeparmdecl, typedefnamedecl, unresolvedusingtypenamedecl'

using_shadowDecl_text = 'constructorusingshadowdecl'

valueDecl_text = 'bindingdecl, declaratordecl, enumconstantdecl, indirectfilddecl'

declaratordecl_text = 'fielddecl, functiondecl, nontypetemplateparmdecl, vardecl'

def add_type_class(key, text):
    if key not in type_class_dict:
        type_class_dict[key] = []
    for i in text.split(','):
        type_class_dict[key].append(i.replace(' ', ''))

add_type_class('stmt', stmt_text)
add_type_class('decl', decl_text)
add_type_class('nameddecl', nameddecl_text)
add_type_class('objccontainerdecl', objccontainerdecl_text)
add_type_class('templatedecl', templateDecl_text)
add_type_class('typedecl', typeDecl_text)
add_type_class('valuedecl', valueDecl_text)
add_type_class('declaratordecl', declaratordecl_text)

    
type_class_dict['asmstmt'] = ['gccasmstmt', 'msasmstmt']
type_class_dict['valuestmt'] = ['attributedstmt', 'expr', 'labelstmt']
type_class_dict['expr'] = ['abstractconditionaloperator', 'addrlabelexpr', 'arrayinitindexexpr', 
                           'arrayinitloopexpr', 'arraysubscriptexpr', 'arraytypetraitexpr', 
                           'astypeexpr', 'atomicexpr', 'binaryoperator', 'blockexpr', 'callexpr',
                          'castexpr', 'cxxconstructexpr']
type_class_dict['callexpr'] = ['cxxmembercallexpr', 'cxxoperatorcallexpr']
type_class_dict['functiondecl'] = ['cxxdeductionguidedecl', 'cxxmethoddecl']
type_class_dict['cxxmethoddecl'] = ['cxxconstructordecl', 'cxxconversiondecl', 'cxxdestructordecl']
type_class_dict['castexpr'] = ['explicitcastexpr']
type_class_dict['explicitcastexpr'] = ['cxxnamedcastexpr']
type_class_dict['cxxnamedcastexpr'] = ['cxxconstcastexpr']

    


# t1, return type; t2, input type list
def type_inheritage(type_class_dict, t1, t2):
#     print(t1, t2)
    if t1 in t2:
        return True
    t1 = t1.lower()
    t2 = [t.lower() for t in t2]
    if t1 in type_class_dict:
        for t in type_class_dict[t1]:
            if type_inheritage(type_class_dict, t, t2):
                return True
    for tt in t2:
        if tt in type_class_dict:
            for t in type_class_dict[tt]:
                if type_inheritage(type_class_dict, t, [t1]):
                    return True
    return False


print(type_inheritage(type_class_dict, 'decl', ['nameddecl']))
print(type_inheritage(type_class_dict, 'Decl', ['ValueDecl']))
print(type_inheritage(type_class_dict, 'NamedDecl', ['Decl']))
print(type_inheritage(type_class_dict, 'TypeDecl', ['ConceptDecl']))
print(type_inheritage(type_class_dict, 'qualtype', ['type']))
# print('StmtDecl'.lower())
```

```python
# Set arguments for node matchers
# for node in node_matcher_list:
for node in gram.api_type_dict['node']:
    node_arg = node + '_arg'
    
    # add empty input
    gram.add_nt(node_arg, ['empty'], node)
                
    # add Matcher<*> node
    gram.add_nt(node_arg, ['Matcher'], node)
                
    matcher_api = []
    
    input_type = gram.api_dict[node].param

    # add traversal matchers as input
    tr_api = []
#     for tr in traversal_matcher_list:
    for tr in gram.api_type_dict['traversal']:
        if gram.api_dict[tr].ret in ['Matcher<*>']:
            matcher_api.append([tr])
            continue
        if type_inheritage(type_class_dict, gram.api_dict[tr].ret, input_type):
            tr_api.append([tr])
    
    for tr in tr_api:
        gram.add_nt(node_arg, tr, node)
    
    # add narrowing matchers as input
    na_api = []
#     for na in narrowing_matcher_list:
    for na in gram.api_type_dict['narrowing']:
        if gram.api_dict[na].ret in ['Matcher<*>']:
            matcher_api.append([na])
            continue
        if type_inheritage(type_class_dict, gram.api_dict[na].ret, input_type):
            na_api.append([na])
    
    for na in na_api:
        gram.add_nt(node_arg, na, node)
        
for m in matcher_api:    
    gram.add_nt('Matcher', m, 'Node')
gram.nt_display()
    
```

```python
# set arguments for narrowing matchers
# for narrowing in narrowing_matcher_list:
for narrowing in gram.api_type_dict['narrowing']:
    narrowing_arg = narrowing + '_arg'
    
    if not gram.api_dict[narrowing].param[0]:
        gram.add_nt(narrowing_arg, ['empty'], narrowing)
        continue
    else:
        input_type = gram.api_dict[narrowing].param
        for inp in input_type:
            if inp in gram.ret_type_dict:
                if 'Matcher<*>' in inp:
                    gram.add_nt(narrowing_arg, ['Node'], narrowing)
                else:
                    node_api = []
                    for nd in gram.api_type_dict['node']:
                        if type_inheritage(type_class_dict, gram.api_dict[nd].ret, input_type):
                            node_api.append([nd])

                    for nd in node_api:
                        gram.add_nt(narrowing_arg, nd, narrowing) 
            else:
                gram.add_nt(narrowing_arg, [inp], narrowing)

gram.nt_display()
            
```

```python
# set arguments for traversal matchers
# for narrowing in narrowing_matcher_list:
for traversal in gram.api_type_dict['traversal']:
    traversal_arg = traversal + '_arg'
    
    if not gram.api_dict[traversal].param[0]:
        gram.add_nt(traversal_arg, ['empty'], traversal)
        continue
    else:
        input_type = gram.api_dict[traversal].param
        for inp in input_type:
            if inp in gram.ret_type_dict:
                if 'Matcher<*>' in inp:
                    gram.add_nt(traversal_arg, ['Node'], traversal)
                else:
                    node_api = []
                    for nd in gram.api_type_dict['node']:
                        if type_inheritage(type_class_dict, gram.api_dict[nd].ret, input_type):
                            node_api.append([nd])

                    for nd in node_api:
                        gram.add_nt(traversal_arg, nd, traversal) 
                    
                    nr_api = []
                    for nr in gram.api_type_dict['narrowing']:
                        if type_inheritage(type_class_dict, gram.api_dict[nr].ret, input_type):
                            nr_api.append([nr])
                    for nr in nr_api:
                        gram.add_nt(traversal_arg, nr, traversal)
            else:
                gram.add_nt(traversal_arg, [inp], traversal)

gram.nt_display()          
```
